```python
loans=pd.read_csv(r'C:\Users\Rayza\Python_Exercises\183813886\pandas\kiva_loans.csv')
```

```python
loans
```

```python
agri_loans=loans[
    (loans['term_in_months']>11)&
    (loans['term_in_months']<25)&
    (loans['loan_amount']<600)&
    (loans['sector'].isin(['Food','Agriculture']))
    ]
```

```python
agri_loans
```

```python
agri_loans.loc[agri_loans['repayment_interval'].isin(['monthly']),
                ['country','partner_id','currency','loan_amount',
                 'funded_amount','sector','activity','term_in_months',
                 'repayment_interval','date']]
```

```python
df=agri_loans.groupby("country").agg(
    avg_loan=('loan_amount','mean'),
    loan_std=('loan_amount','std'),
    avg_fund=('funded_amount','mean'),
    fund_std=('funded_amount','std'),
    avg_term_in_months=('term_in_months','mean'),
).reset_index()
df
```

```python
loansbycountry=df.round(2)
loansbycountry
```

```python
loansbycountry['investment'] = loansbycountry.apply(lambda row: np.sum(row['avg_loan'] + row['avg_fund']), axis=1)
```
